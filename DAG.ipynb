```python
import random
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
num_plots = 30
f = plt.figure()

for node_number in range(num_plots):

    graph.add_node(node_number, Position=(random.randrange(0, 100), random.randrange(0, 100)))
    graph.add_edge(node_number, random.choice(list(graph.nodes())))
    nx.draw(graph, pos=nx.get_node_attributes(graph,'Position'),ax=f.add_subplot(111))
    f.savefig("ani/{}.png".format(node_number))

```

```python
import os
import imageio

png_dir = 'ani'
images = []

# for f in os.listdir(png_dir):
#     os.remove(os.path.join(png_dir, f))
    
    
for node_number in range(num_plots):
        file_path = os.path.join(png_dir, str(node_number)+".png")
        images.append(imageio.imread(file_path))
imageio.mimsave('movie.gif', images,duration=0.2)
```
