```python
import random
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.Graph()
num_plots = 100
f = plt.figure(figsize=(8, 6), dpi=80)
pos=2

end_pos = []

for files in os.listdir('ani/'):
    os.remove(os.path.join('ani/', files))
    
for node_number in range(num_plots):
    
    
    pos += 30
    rand = random.randrange(-150, 55)




    if node_number == 0:
        graph.add_node(node_number, Position=(0, 0))  
    
        graph.add_edge(node_number, 0)



    elif node_number>0 and node_number <=14 :

        graph.add_node(node_number, Position=(0, pos))  
    
        graph.add_edge(node_number, node_number-1)
        
        end_pos.append(node_number)
 
    
        
        
        
    else:
        graph.add_node(node_number, Position=(pos, rand))  
        
        # initializing the value of n
        n = 3

        # printing n elements from list
        links = random.sample(end_pos, n)
        
#         print(end_pos)
        for i in range(len(links)):
            graph.add_edge(node_number,links[i])
            end_pos.remove(links[i])
            end_pos.append(node_number)

        
        
        
  
    nx.draw(graph, arrowsize=200, arrowstyle='fancy', pos=nx.get_node_attributes(graph,'Position'),ax=f.add_subplot(111))
    f.savefig("ani/{}.png".format(node_number))


import os
import imageio
from IPython.display import Image
png_dir = 'ani'
images = []

# for f in os.listdir(png_dir):
#     os.remove(os.path.join(png_dir, f))
    
    
for node_number in range(num_plots):
        file_path = os.path.join(png_dir, str(node_number)+".png")
        images.append(imageio.imread(file_path))
imageio.mimsave('movie2.gif', images,duration=0.041)

Image(filename='movie2.gif')
```
